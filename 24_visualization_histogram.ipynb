### Histogram

#### Histogram แสดงการกระจายตัว(distribution) ของข้อมูลที่เราสนใจ
#### แกน x คือข้อมูลหรือสิ่งที่เราสนใจ โดยจะถูกแบ่งออกเป็นช่วงๆ 
#### แกน Y คือความถี่ในการเกิดข้อมูลนั้นๆหรือสิ่งที่เราสนใจนั้นๆ
##### https://en.wikipedia.org/wiki/Histogram

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
#draw displot
sns.distplot(df['cty']) #สร้าง hittogram
```

```python
#kde
sns.distplot(df['cty'],kde = False) #ลบเส้น kde
```

```python
sns.distplot(df['cty'], kde=False)
plt.axvline(x=df.cty.mean(),color = 'red',linestyle = '--') 
plt.title(f"mean = {df.cty.mean():.2f}, sd = {df.cty.std():.2f}") #กำหนด title
```

```python
#draw 2 displot : 1 figure
sns.distplot(df['cty'],kde=False,label = 'City')
sns.distplot(df['hwy'],kde=False,label = 'higthway')
plt.legend()
```

```python
#draw 2 displot 
sns.distplot(df['cty'],kde=False,label = 'City')
plt.show()
sns.distplot(df['hwy'],kde=False,label = 'higthway')
plt.show()
```

```python
fig, ax = plt.subplots(1,2) 
print(ax.shape)
sns.distplot(df['cty'],kde=False,label = 'City',ax = ax[0])
sns.distplot(df['hwy'],kde=False,label = 'higthway',ax = ax[1])
```

```python
#sharey
fig, ax = plt.subplots(1,2,sharey = True) #แกน Y
print(ax.shape)
sns.distplot(df['cty'],kde=False,label = 'City',ax = ax[0])
sns.distplot(df['hwy'],kde=False,label = 'higthway',ax = ax[1])
```

```python
#sharey
fig, ax = plt.subplots(1,2,sharex = True, sharey = True) #แกน X, Y
print(ax.shape)
sns.distplot(df['cty'],kde=False,label = 'City',ax = ax[0])
sns.distplot(df['hwy'],kde=False,label = 'higthway',ax = ax[1])
```
