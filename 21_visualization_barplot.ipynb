## barplot
**A bar plot or bar chart is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline 
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
#barplot = y is number, x = cotigory ข้อมูลที่เป็นกลุ่ม
sns.barplot(x='gear', y='cty', data=df)
```

```python
sns.barplot(x='gear', y='cty', data=df, hue='drv')
```

```python
df.groupby('drv')['cty'].describe()
```

```python
sns.barplot(x='drv', y='cty', data=df,ci=None) #ci = ค่าช่วงความเชื่อมั่นที่เป็นขีดๆด้านบน None = ไม่แสดง  
```

```python
# estimator = plot as y
sns.barplot(x='drv', y='cty', data=df,ci=None, estimator=max)
```

```python
sns.barplot(x='drv', y='cty', data=df,ci=None, estimator=min)
```

```python
sns.barplot(x='drv', y='cty', data=df,ci=None, estimator=np.median) #np.median = mean
```

```python
avg=df.cty.mean() 
sns.barplot(x='drv', y='cty', data=df)
plt.axhline(y=avg, color='red', linestyle='--')
```

```python
avg=df.cty.mean()
sns.barplot(x='drv', y='cty', data=df, 
            hue='gear',
            palette='Set1', #color as graph 
            ci=None,
            ) 
plt.axhline(y=avg, color='red', linestyle='--')
# https://seaborn.pydata.org/generated/seaborn.color_palette.html
```

```python
avg=df.cty.mean()
plt.figure(figsize=(10,7)) #resize graph
sns.barplot(x='drv', y='cty', data=df, 
            hue='gear',
            palette='Set1', 
            ci=None,
            )
plt.axhline(y=avg, color='red', linestyle='--')
```

```python
plt.figure(figsize=(8,6))  
sns.barplot(x='drv', y='cty', data=df,
            hue='gear', 
            ci=None,
            palette=['#dd0000', 'plum']) #เป็นการระบุค่าสีลงไปเลย # palette  https://matplotlib.org/3.1.0/gallery/color/named_colors.html
```

```python
plt.figure(figsize=(8,6)) 
sns.barplot(x='drv', y='cty', data=df,
            hue='gear', 
            ci=None,
            palette=['#dd0000', 'plum'])  # palette  https://matplotlib.org/3.1.0/gallery/color/named_colors.html
plt.xlabel('Name drv') #แสดงชื่อ หรือเปลี่ยนชื่อของแกน x
plt.ylabel('city miles per gallon') #แสดงชื่อ หรือเปลี่ยนชื่อของแกน y
plt.title('fuel efficiency by drive wheels') 
```

```python
sns.factorplot(x='drv', y='cty', data=df,ci=None, #factorplot = plot graph หลายๆ ติดต่อกัน
               hue="gear", 
               col="year",
               kind="bar",
               height=4, aspect=1);
```

```python
sns.factorplot(x='drv', y='cty', data=df[df['class'].str.match('compact|suv|midsize')],
               hue='gear', 
               row='year',
               col='class',
               kind='bar',
               ci=None,
               palette=['deepskyblue', 'tomato'],
               height=3, aspect=1);
```
