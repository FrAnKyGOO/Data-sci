### Boxplot

#### Boxplot แสดง ค่ากลาง ค่าการกระจาย สัดส่วนข้อมูลที่มากหรือน้อยกว่าค่ากลาง 
#### รวมทั้งข้อมูลที่ผิดปกติ(Outlier) เป็นข้อมูลที่อยู่ห่างจากกลุ่มมากๆ 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
url='https://github.com/mathawanup/learn_dataset/raw/master/car_mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
sns.boxplot(data=df[['cty']]) #data is list
```

```python
#draw 2 boxplot : 1 figure
sns.boxplot(data=df[['cty','hwy']])
```

```python
#cty by drv
sns.boxplot(x='drv', y='cty' ,data = df) 
```

```python
sns.boxplot() #reverse
sns.boxplot(y='drv', x='cty' ,data = df) 
```

```python
sns.boxplot(y='drv', x='cty',data=df)
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#hue year
sns.boxplot(y='drv', x='cty',data=df, hue = 'year') #เปรียบเทียบกันในปี
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#figsize
plt.figure(figsize=(10,4)) #ขยายให้ใหญ่ขึ้น
sns.boxplot(y='drv', x='cty',data=df, hue = 'year') #เปรียบเทียบกันในปี
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#palette
plt.figure(figsize=(10,4)) #ขยายให้ใหญ่ขึ้น
sns.boxplot(y='drv', x='cty',data=df, hue = 'year',palette='Set1') #เปรียบเทียบกันในปี,setของสี
plt.ylabel('')
plt.xlabel('city miles per gallon')
plt.title('miles per gallon by driving system')
```

```python
#draw 2 plot
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              col='year') 
```

```python
sns.catplot(y='drv', x='cty', data=df,
              kind='box',
              row='year',
               col='gear')
```
