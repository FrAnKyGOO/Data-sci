**การสำรวจข้อมูลเบื้องต้น 3 (Basic Data exploration)**

**การสำรวจแถวและคอลัมน์ใน pandas DataFrame ที่มี label index ด้วยคำสั่ง loc**

```python
import pandas as pd
```

### Select column

```python
url='https://raw.githubusercontent.com/mathawanup/master_data/master/score_a.tsv'
df=pd.read_table(url, index_col='studentid') # กำหนด studentid ให้เป็น index_col 
df
```

```python
#ดูว่ามี columns อะไรบ้าง
df.columns 
```

```python
#เปรียบเทียบตัวอักษรทีละตัว เหมือน = true , ไม่เหมือน = false
df.columns.str.contains('eng')
```

```python
# ถ้า index เป็นแบบ lable จะใช้ loc[] ในการเข้าถึงทุกแถว
df.loc[(df.math1>=5)] 
```

```python
df.loc[:, df.columns.str.contains('eng')] 
```

```python
 # eng and sci
df.loc[:, df.columns.str.contains('eng|sci')] 
```

```python
df.columns.str.match('(math|sci)[1|3]') #ชื่อ col ตามเงื่อนไข
```

```python
df.loc[:,df.columns.str.match('(math|sci)[1|3]')] #ชื่อ col ตามเงื่อนไข ทุกๆแถว
```

### Select row

**1.single label**

```python
#ระบุ studentid	
df.loc['C215']
```

```python
df.loc[['C215']]
```

```python
#ระบุ row แสดงเฉพาะ sci1.
df.loc[['C215'],'sci1']
```

**2.list of labels**

```python
 #ระบุ list of row
df.loc[['C215', 'B344']]
```

```python
 #ระบุ list of rowand col
df.loc[['C215', 'B344'],['math1','math2','math3']]
```

**3.slice object**

```python
# ระบุ list of row แบบ slice
df.loc['C215':'A312'] #[เริ่ม:ถึงสิ้นสุด-1]
```

```python
# แบบ slice row,col
df.loc['C215':'A312','math1':'sci2']
```

**4.boolean array**

```python
#แสดงเฉพาะ index
df.index
```

```python
#หาตัวที่ขึ้นต้นตามที่กำหนด
df.index.str.startswith('B') #index  startswith('B')
```

```python
#index  startswith('B')  show in dataFrame
df.loc[df.index.str.startswith('B')]
```

```python
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')]
```

```python
# sum แต่ละ row ตามเงื่อนไข
# axis = 1 ----> รวมแต่ละ row
# axis = 0 ----> รวมแต่ละ col
df.loc[df.index.str.startswith('B'), df.columns.str.endswith('3')].sum(axis=1)
```
