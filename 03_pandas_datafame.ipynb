```python
# https://github.com/mathawanup/learn_dataset/blob/master/mpg.csv
df=pd.read_csv('https://raw.githubusercontent.com/mathawanup/learn_dataset/master/mpg.csv')
df.head()
```

```python
df['cty_kml'] = df['cty'] * .425143707
df.head()
```

```python
df['cty_kml'] = df['cty_kml'].round(1)
df.head()
```

```python
df['trans_am'] = df.trans.str[0] # สามารถใช้ . ตามด้วยชื่อ colunms เพื่ออ้างอิงได้
df.head()
```

```python
filepath = 'E:\workshop\data_mpg.csv'
df.to_csv(filepath) # save dataFrame เป็น csv
```

```python
filepath = 'E:\workshop\data_mpg.xlsx' #  $ conda install openpyxl  ติดตั้ง Anaconda prompt
writer=pd.ExcelWriter(filepath) # save dataFrame เป็น Excel
df.to_excel(writer)
writer.save()
```
